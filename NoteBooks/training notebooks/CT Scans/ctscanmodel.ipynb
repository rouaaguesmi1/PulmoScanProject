```python
# -*- coding: utf-8 -*-
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader, random_split, Subset
import time
import copy
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm # Use notebook version if in Kaggle/Jupyter
# from tqdm import tqdm # Use this if running a standard .py script
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, f1_score

# Ensure reproducibility (optional, but good practice)
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False # False for determinism, True for performance if input size doesn't change

print("PyTorch Version:", torch.__version__)

# --- Configuration ---
# Define the dataset path
data_dir = r'/kaggle/input/ct-scans/Chest CT-Scan images Dataset'

# Set image dimensions and batch size
IMG_HEIGHT, IMG_WIDTH = 224, 224
BATCH_SIZE = 64       # Increase batch size for powerful GPU (adjust if OOM)
NUM_EPOCHS = 75       # Increase epochs, rely on early stopping
LEARNING_RATE = 1e-3  # Higher initial LR for training from scratch
WEIGHT_DECAY = 1e-4   # Standard weight decay
EARLY_STOPPING_PATIENCE = 15 # Increased patience for longer training
LABEL_SMOOTHING = 0.1 # Use label smoothing
GRADIENT_CLIP_NORM = 1.0 # Max norm for gradient clipping

MODEL_SAVE_PATH = '/kaggle/working/best_custom_sota_model.pth'
CHECKPOINT_SAVE_PATH = '/kaggle/working/training_checkpoint.pth'

# Detect if we have a GPU available
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
if device.type == 'cuda':
    print(f"Device name: {torch.cuda.get_device_name(0)}")

# --- Data Loading and Preprocessing ---

# Data augmentation and normalization for training
# Keep strong augmentations
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop((IMG_HEIGHT, IMG_WIDTH), scale=(0.8, 1.0)),
    transforms.RandomRotation(15),
    transforms.RandomHorizontalFlip(),
    # transforms.ColorJitter(brightness=0.1, contrast=0.1), # Less relevant for CT
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]) # Standard ImageNet normalization
])

# Just resize and normalization for validation/testing
val_transforms = transforms.Compose([
    transforms.Resize((IMG_HEIGHT + 32, IMG_WIDTH + 32)), # Resize slightly larger
    transforms.CenterCrop((IMG_HEIGHT, IMG_WIDTH)),      # Then center crop
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Create the full dataset using ImageFolder
try:
    full_dataset = datasets.ImageFolder(data_dir)
    classes = full_dataset.classes
    num_classes = len(classes)
    print("Classes found:", classes)
    print("Class indices:", full_dataset.class_to_idx)
except FileNotFoundError:
    print(f"Error: Data directory not found at {data_dir}")
    exit() # Or handle appropriately

# Split the dataset into training and validation (80% train, 20% validation)
val_split = 0.2
num_total = len(full_dataset)
indices = list(range(num_total))
split = int(np.floor(val_split * num_total))
np.random.shuffle(indices) # Shuffle indices before splitting

train_indices, val_indices = indices[split:], indices[:split]

# Create subset datasets with appropriate transforms
# Important: Apply transforms *after* splitting by using separate datasets or custom logic
train_dataset = datasets.ImageFolder(data_dir, transform=train_transforms)
val_dataset = datasets.ImageFolder(data_dir, transform=val_transforms)

train_subset = Subset(train_dataset, train_indices)
val_subset = Subset(val_dataset, val_indices)

# Create DataLoaders
num_workers = os.cpu_count()//2 if os.cpu_count() and os.cpu_count() > 2 else 2 # Use a few workers
train_loader = DataLoader(train_subset, batch_size=BATCH_SIZE, shuffle=True, num_workers=num_workers, pin_memory=True if device.type=='cuda' else False)
val_loader = DataLoader(val_subset, batch_size=BATCH_SIZE, shuffle=False, num_workers=num_workers, pin_memory=True if device.type=='cuda' else False)

dataloaders = {'train': train_loader, 'val': val_loader}
dataset_sizes = {'train': len(train_subset), 'val': len(val_subset)}

print(f"Training samples: {dataset_sizes['train']}, Validation samples: {dataset_sizes['val']}")
print(f"Num workers: {num_workers}")

# --- Custom SOTA-Inspired Model Architecture ---

class ConvBNReLU(nn.Sequential):
    def __init__(self, in_planes, out_planes, kernel_size=3, stride=1, groups=1):
        padding = (kernel_size - 1) // 2
        super(ConvBNReLU, self).__init__(
            nn.Conv2d(in_planes, out_planes, kernel_size, stride, padding, groups=groups, bias=False),
            nn.BatchNorm2d(out_planes),
            nn.ReLU6(inplace=True) # ReLU6 is common in efficient architectures
        )

class InvertedResidual(nn.Module):
    def __init__(self, inp, oup, stride, expand_ratio):
        super(InvertedResidual, self).__init__()
        self.stride = stride
        assert stride in [1, 2]

        hidden_dim = int(round(inp * expand_ratio))
        self.use_res_connect = self.stride == 1 and inp == oup

        layers = []
        if expand_ratio != 1:
            # pw
            layers.append(ConvBNReLU(inp, hidden_dim, kernel_size=1))
        layers.extend([
            # dw
            ConvBNReLU(hidden_dim, hidden_dim, stride=stride, groups=hidden_dim),
            # pw-linear
            nn.Conv2d(hidden_dim, oup, 1, 1, 0, bias=False),
            nn.BatchNorm2d(oup),
        ])
        self.conv = nn.Sequential(*layers)

    def forward(self, x):
        if self.use_res_connect:
            return x + self.conv(x)
        else:
            return self.conv(x)

class CustomSotaCNN(nn.Module):
    def __init__(self, num_classes=4, width_mult=1.0):
        super(CustomSotaCNN, self).__init__()
        block = InvertedResidual
        input_channel = 32
        last_channel = 1280 # Common last channel size

        # Stem: Similar to ResNet/EfficientNet start
        self.stem = nn.Sequential(
            ConvBNReLU(3, input_channel, kernel_size=3, stride=2),
            # nn.MaxPool2d(kernel_size=3, stride=2, padding=1) # Optional extra pooling
        )

        # Building blocks (inverted residuals)
        # Config: expansion factor, output channels, number of repeats, stride
        inverted_residual_setting = [
            # t, c, n, s
            [1, 16, 1, 1],
            [6, 24, 2, 2],
            [6, 40, 2, 2],
            [6, 80, 3, 2],
            [6, 112, 3, 1],
            [6, 192, 4, 2],
            [6, 320, 1, 1],
        ]

        features = []
        # Make channels scale with width_mult
        input_channel = int(input_channel * width_mult)

        # Build stages
        for t, c, n, s in inverted_residual_setting:
            output_channel = int(c * width_mult)
            for i in range(n):
                stride = s if i == 0 else 1
                features.append(block(input_channel, output_channel, stride, expand_ratio=t))
                input_channel = output_channel
        self.features = nn.Sequential(*features)

        # Building last several layers
        last_conv_output_channel = int(last_channel * max(1.0, width_mult))
        self.final_conv = ConvBNReLU(input_channel, last_conv_output_channel, kernel_size=1)

        # Building classifier
        self.pool = nn.AdaptiveAvgPool2d((1, 1))
        self.classifier = nn.Sequential(
            nn.Dropout(0.3), # Slightly higher dropout
            nn.Linear(last_conv_output_channel, num_classes),
        )

        # Weight initialization (important when training from scratch)
        self._initialize_weights()

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out')
                if m.bias is not None:
                    nn.init.zeros_(m.bias)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.ones_(m.weight)
                nn.init.zeros_(m.bias)
            elif isinstance(m, nn.Linear):
                nn.init.normal_(m.weight, 0, 0.01)
                nn.init.zeros_(m.bias)

    def forward(self, x):
        x = self.stem(x)
        x = self.features(x)
        x = self.final_conv(x)
        x = self.pool(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x

# Instantiate the model
model = CustomSotaCNN(num_classes=num_classes, width_mult=1.0) # Start with width=1.0
model = model.to(device)

# --- Loss Function and Optimizer ---

# Use Label Smoothing
criterion = nn.CrossEntropyLoss(label_smoothing=LABEL_SMOOTHING)

# Use AdamW optimizer
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

# Use Cosine Annealing LR scheduler
scheduler = lr_scheduler.CosineAnnealingLR(optimizer, T_max=NUM_EPOCHS, eta_min=LEARNING_RATE/100)

# --- Training Function ---

def train_model(model, criterion, optimizer, scheduler, num_epochs=25, patience=10, gradient_clip_norm=1.0):
    since = time.time()

    best_model_wts = copy.deepcopy(model.state_dict())
    best_val_loss = float('inf')
    best_val_f1 = 0.0
    epochs_no_improve = 0
    start_epoch = 0

    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': [], 'val_f1_macro': [], 'val_f1_weighted': [], 'lr': []}

    # --- Load checkpoint if exists ---
    if os.path.exists(CHECKPOINT_SAVE_PATH):
        print(f"Loading checkpoint from {CHECKPOINT_SAVE_PATH}")
        checkpoint = torch.load(CHECKPOINT_SAVE_PATH, map_location=device)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        scheduler.load_state_dict(checkpoint['scheduler_state_dict'])
        start_epoch = checkpoint['epoch'] + 1
        best_val_loss = checkpoint['best_val_loss']
        best_val_f1 = checkpoint.get('best_val_f1', 0.0) # Handle older checkpoints
        history = checkpoint['history']
        epochs_no_improve = checkpoint.get('epochs_no_improve', 0)
        best_model_wts = copy.deepcopy(model.state_dict()) # Re-copy best weights after loading
        print(f"Resuming training from epoch {start_epoch}")


    for epoch in range(start_epoch, num_epochs):
        print(f'\nEpoch {epoch}/{num_epochs - 1}')
        print('-' * 15)

        # Each epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # Set model to training mode
            else:
                model.eval()   # Set model to evaluate mode

            running_loss = 0.0
            running_corrects = 0
            all_labels = []
            all_preds = []

            # Use tqdm for progress bar
            data_iterator = tqdm(dataloaders[phase], desc=f"{phase.capitalize()} Epoch {epoch}", leave=False)

            for inputs, labels in data_iterator:
                inputs = inputs.to(device)
                labels = labels.to(device)

                # Zero the parameter gradients
                optimizer.zero_grad()

                # Forward
                # Track history if only in train
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    # Backward + optimize only if in training phase
                    if phase == 'train':
                        loss.backward()
                        # Gradient Clipping
                        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=gradient_clip_norm)
                        optimizer.step()

                # Statistics
                batch_loss = loss.item()
                batch_corrects = torch.sum(preds == labels.data).item()
                running_loss += batch_loss * inputs.size(0)
                running_corrects += batch_corrects

                # Collect labels/preds for F1/Confusion Matrix in validation phase
                if phase == 'val':
                    all_labels.extend(labels.cpu().numpy())
                    all_preds.extend(preds.cpu().numpy())

                # Update tqdm progress bar
                data_iterator.set_postfix(loss=batch_loss, acc=batch_corrects/inputs.size(0))

            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects / dataset_sizes[phase]

            current_lr = optimizer.param_groups[0]['lr']
            print(f'{phase.capitalize()} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f} LR: {current_lr:.6f}')

            # Store history
            history['lr'].append(current_lr)
            if phase == 'train':
                history['train_loss'].append(epoch_loss)
                history['train_acc'].append(epoch_acc)
                # Step the scheduler after the training phase
                scheduler.step()
            else:
                history['val_loss'].append(epoch_loss)
                history['val_acc'].append(epoch_acc)

                # Calculate F1 Scores for validation phase
                epoch_f1_macro = f1_score(all_labels, all_preds, average='macro', zero_division=0)
                epoch_f1_weighted = f1_score(all_labels, all_preds, average='weighted', zero_division=0)
                history['val_f1_macro'].append(epoch_f1_macro)
                history['val_f1_weighted'].append(epoch_f1_weighted)
                print(f'Val F1 Macro: {epoch_f1_macro:.4f}, Val F1 Weighted: {epoch_f1_weighted:.4f}')

                # Early Stopping and Best Model Saving (based on validation loss)
                if epoch_loss < best_val_loss:
                    print(f"Validation loss decreased ({best_val_loss:.4f} --> {epoch_loss:.4f}). Saving model...")
                    best_val_loss = epoch_loss
                    best_val_f1 = epoch_f1_macro # Store best F1 corresponding to best loss
                    best_model_wts = copy.deepcopy(model.state_dict())
                    torch.save(best_model_wts, MODEL_SAVE_PATH) # Save only best weights separately
                    epochs_no_improve = 0

                    # Save comprehensive checkpoint for resuming
                    torch.save({
                        'epoch': epoch,
                        'model_state_dict': best_model_wts, # Save best weights here too
                        'optimizer_state_dict': optimizer.state_dict(),
                        'scheduler_state_dict': scheduler.state_dict(),
                        'best_val_loss': best_val_loss,
                        'best_val_f1': best_val_f1,
                        'history': history,
                        'epochs_no_improve': epochs_no_improve,
                        'class_to_idx': dataloaders['train'].dataset.dataset.class_to_idx # Save class mapping
                    }, CHECKPOINT_SAVE_PATH)

                else:
                    epochs_no_improve += 1
                    print(f"Validation loss did not improve for {epochs_no_improve} epoch(s).")

                     # Save checkpoint periodically even if not improving
                    if epoch % 5 == 0: # Save every 5 epochs regardless
                         print("Saving periodic checkpoint...")
                         torch.save({
                            'epoch': epoch,
                            'model_state_dict': model.state_dict(), # Current weights
                            'optimizer_state_dict': optimizer.state_dict(),
                            'scheduler_state_dict': scheduler.state_dict(),
                            'best_val_loss': best_val_loss, # Keep track of the best loss so far
                            'best_val_f1': best_val_f1,
                            'history': history,
                            'epochs_no_improve': epochs_no_improve,
                            'class_to_idx': dataloaders['train'].dataset.dataset.class_to_idx
                         }, CHECKPOINT_SAVE_PATH)


                if epochs_no_improve >= patience:
                    print(f"Early stopping triggered after {epoch + 1} epochs.")
                    model.load_state_dict(best_model_wts) # Load best model weights
                    time_elapsed = time.time() - since
                    print(f'Training complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
                    print(f'Best val Loss: {best_val_loss:.4f}, Best val F1 Macro: {best_val_f1:.4f}')
                    return model, history # Exit training loop


    time_elapsed = time.time() - since
    print(f'Training complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
    print(f'Best val Loss: {best_val_loss:.4f}, Best val F1 Macro: {best_val_f1:.4f}')

    # Load best model weights at the end
    model.load_state_dict(best_model_wts)
    return model, history

# --- Train the Model ---
try:
    model, history = train_model(model, criterion, optimizer, scheduler,
                                 num_epochs=NUM_EPOCHS, patience=EARLY_STOPPING_PATIENCE,
                                 gradient_clip_norm=GRADIENT_CLIP_NORM)
except Exception as e:
    print(f"An error occurred during training: {e}")
    # Potentially try loading the last checkpoint if model/history are None
    if os.path.exists(CHECKPOINT_SAVE_PATH) and ('model' not in locals() or model is None):
         print("Attempting to load model from last checkpoint for evaluation.")
         checkpoint = torch.load(CHECKPOINT_SAVE_PATH, map_location=device)
         model = CustomSotaCNN(num_classes=len(checkpoint['class_to_idx']), width_mult=1.0) # Re-init model structure
         model.load_state_dict(checkpoint['model_state_dict'])
         model.to(device)
         history = checkpoint.get('history', None) # Get history if available
    else:
        print("Could not recover model or history.")
        model = None
        history = None

# --- Evaluate Best Model and Visualize Metrics ---

def evaluate_and_visualize(model, dataloader, device, class_names, history=None):
    if model is None:
        print("Model is not available for evaluation.")
        return

    if history:
         # --- Plot Training History ---
        print("\nPlotting Training History...")
        fig, axes = plt.subplots(2, 2, figsize=(18, 10))
        fig.suptitle('Training Metrics')

        # Accuracy
        axes[0, 0].plot(history['train_acc'], label='Train Accuracy')
        axes[0, 0].plot(history['val_acc'], label='Validation Accuracy')
        axes[0, 0].set_title('Model Accuracy')
        axes[0, 0].set_ylabel('Accuracy')
        axes[0, 0].set_xlabel('Epoch')
        axes[0, 0].legend(loc='best')
        axes[0, 0].grid(True)

        # Loss
        axes[0, 1].plot(history['train_loss'], label='Train Loss')
        axes[0, 1].plot(history['val_loss'], label='Validation Loss')
        axes[0, 1].set_title('Model Loss')
        axes[0, 1].set_ylabel('Loss')
        axes[0, 1].set_xlabel('Epoch')
        axes[0, 1].legend(loc='best')
        axes[0, 1].grid(True)

        # F1 Score
        axes[1, 0].plot(history['val_f1_macro'], label='Validation F1 Macro')
        axes[1, 0].plot(history['val_f1_weighted'], label='Validation F1 Weighted')
        axes[1, 0].set_title('Validation F1 Score')
        axes[1, 0].set_ylabel('F1 Score')
        axes[1, 0].set_xlabel('Epoch')
        axes[1, 0].legend(loc='best')
        axes[1, 0].grid(True)

        # Learning Rate
        axes[1, 1].plot(history['lr'], label='Learning Rate')
        axes[1, 1].set_title('Learning Rate Schedule')
        axes[1, 1].set_ylabel('Learning Rate')
        axes[1, 1].set_xlabel('Epoch')
        axes[1, 1].legend(loc='best')
        axes[1, 1].grid(True)

        plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
        plt.show()
    else:
        print("History data not available for plotting.")


    # --- Final Evaluation on Validation Set ---
    print("\nEvaluating best model on validation set...")
    model.eval()
    all_labels = []
    all_preds = []

    with torch.no_grad():
        for inputs, labels in tqdm(dataloader, desc="Final Evaluation"):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())

    # --- Classification Report ---
    print("\nClassification Report:")
    report = classification_report(all_labels, all_preds, target_names=class_names, digits=4, zero_division=0)
    print(report)

    # --- Confusion Matrix ---
    print("\nConfusion Matrix:")
    cm = confusion_matrix(all_labels, all_preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)

    fig, ax = plt.subplots(figsize=(8, 8))
    disp.plot(cmap=plt.cm.Blues, ax=ax, xticks_rotation='vertical')
    plt.title('Confusion Matrix')
    plt.tight_layout()
    plt.show()

# Run evaluation if model training finished (or model was loaded)
if model is not None:
    # Load the best weights explicitly before final evaluation
    if os.path.exists(MODEL_SAVE_PATH):
        print(f"\nLoading best model weights from {MODEL_SAVE_PATH} for final evaluation.")
        model.load_state_dict(torch.load(MODEL_SAVE_PATH, map_location=device))
    else:
        print("\nWarning: Best model file not found. Evaluating with the current model state (potentially the last epoch's weights or loaded checkpoint).")

    evaluate_and_visualize(model, val_loader, device, classes, history)
else:
    print("\nSkipping final evaluation as the model could not be trained or loaded.")


print("\n--- Script Finished ---")
```
