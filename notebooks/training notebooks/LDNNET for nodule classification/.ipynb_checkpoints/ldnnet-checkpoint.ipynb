```python
# --- Helper Functions (Mostly unchanged, added PyTorch specifics) ---

def load_itk_image(filename):
    """Loads a .mhd/.raw image or DICOM series using SimpleITK."""
    # ... (same as before) ...
    if os.path.isdir(filename):
        reader = sitk.ImageSeriesReader()
        dicom_names = reader.GetGDCMSeriesFileNames(filename)
        if not dicom_names:
             raise FileNotFoundError(f"No DICOM files found in directory: {filename}")
        reader.SetFileNames(dicom_names)
        itkimage = reader.Execute()
    else:
        itkimage = sitk.ReadImage(filename)
    return itkimage


def get_pixels_hu(itkimage):
    """Converts ITK image pixels to Hounsfield Units (HU)."""
    # ... (same as before) ...
    image = sitk.GetArrayFromImage(itkimage) # z, y, x
    intercept = itkimage.GetMetaData('RescaleIntercept') if 'RescaleIntercept' in itkimage.GetMetaDataKeys() else -1024
    slope = itkimage.GetMetaData('RescaleSlope') if 'RescaleSlope' in itkimage.GetMetaDataKeys() else 1
    image = image.astype(np.int16)
    try:
       intercept = float(intercept)
       slope = float(slope)
    except ValueError:
       print(f"Warning: Could not convert slope/intercept to float. Using defaults (-1024/1)")
       intercept = -1024.0
       slope = 1.0
    if slope != 1:
        image = slope * image.astype(np.float64)
    image = image.astype(np.int16) + np.int16(intercept)
    return np.array(image, dtype=np.int16)

def resample_image(image, spacing, new_spacing=[1.0, 1.0, 1.0]):
    """Resamples image to new spacing using scipy.ndimage.zoom."""
    # ... (same as before) ...
    resize_factor = spacing / new_spacing
    new_real_shape = image.shape * resize_factor
    new_shape = np.round(new_real_shape)
    real_resize_factor = new_shape / image.shape
    # new_spacing = spacing / real_resize_factor # Actual new spacing

    # Order=1 is bilinear interpolation
    image = zoom(image, real_resize_factor, mode='nearest', order=1)

    return image # Return only image for simplicity here


def resize_scan(image, target_shape):
    """Resizes a 3D image to a target shape."""
    factors = np.array(target_shape) / np.array(image.shape)
    # Using order=1 (bilinear) for resizing
    resized_image = zoom(image, factors, mode='nearest', order=1)
    return resized_image

def normalize(image, min_bound=-1000.0, max_bound=400.0):
    """Normalizes pixel values to the range [0, 1]."""
    # ... (same as before) ...
    image = (image - min_bound) / (max_bound - min_bound)
    image[image > 1] = 1.
    image[image < 0] = 0.
    return image.astype(np.float32)


def world_to_voxel(world_coords, origin, spacing):
    """Converts world coordinates (mm) to voxel coordinates."""
    # ... (same as before) ...
    stretched_voxel_coords = np.absolute(world_coords - origin)
    voxel_coords = stretched_voxel_coords / spacing
    return tuple(np.round(voxel_coords).astype(int))


def extract_patch(image, coord_voxel, patch_shape):
    """Extracts a 3D patch centered at coord_voxel."""
    # ... (Ensure patch_shape is D, H, W) ...
    patch_half = np.array(patch_shape) // 2
    coord_start = np.array(coord_voxel) - patch_half # Use np.array for element-wise op
    coord_end = coord_start + np.array(patch_shape) # End is exclusive in slicing

    pad_dims = []
    for i in range(3):
        pad_before = max(0, -coord_start[i])
        pad_after = max(0, coord_end[i] - image.shape[i])
        pad_dims.append((pad_before, pad_after))

    slice_start = np.maximum(0, coord_start)
    slice_end = np.minimum(image.shape, coord_end)

    patch = image[slice_start[0]:slice_end[0],
                  slice_start[1]:slice_end[1],
                  slice_start[2]:slice_end[2]]

    if any(p[0] > 0 or p[1] > 0 for p in pad_dims):
        patch = np.pad(patch, pad_dims, mode='constant', constant_values=0)

    # Final check - Sometimes rounding leads to off-by-one, simple center crop/pad fallback
    if patch.shape != patch_shape:
         # print(f"Warning: Patch shape correction needed ({patch.shape} vs {patch_shape})")
         current_shape = np.array(patch.shape)
         target_shape = np.array(patch_shape)
         diff = target_shape - current_shape
         
         pad_before = diff // 2
         pad_after = diff - pad_before
         
         pad_width = list(zip(np.maximum(0, pad_before), np.maximum(0, pad_after)))
         
         crop_before = np.maximum(0, -pad_before)
         crop_after = np.maximum(0, -pad_after)
         
         slices = tuple(slice(crop_before[i], current_shape[i] - crop_after[i]) for i in range(3))
         
         patch = patch[slices]
         patch = np.pad(patch, pad_width, mode='constant', constant_values=0)

         if patch.shape != patch_shape:
             print(f"ERROR: Patch shape correction failed. Final shape: {patch.shape}")
             return None
    return patch
```

```python
# --- PyTorch Dataset Classes ---

class LUNAPatchDataset(Dataset):
    def __init__(self, patches, labels):
        self.patches = patches # Should be numpy array (N, D, H, W)
        self.labels = labels   # Should be numpy array (N,)
        if not isinstance(self.patches, np.ndarray):
             self.patches = np.array(self.patches)
        if not isinstance(self.labels, np.ndarray):
             self.labels = np.array(self.labels)


    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        patch = self.patches[idx]
        label = self.labels[idx]

        # Add channel dimension (N, C, D, H, W) - PyTorch standard
        patch_tensor = torch.from_numpy(patch).float().unsqueeze(0)
        label_tensor = torch.tensor(label).float() # Use float for BCEWithLogitsLoss

        return patch_tensor, label_tensor

class DSBScanDataset(Dataset):
    def __init__(self, patient_ids, labels_df, data_path, target_size, target_spacing, transform=None):
        self.patient_ids = patient_ids
        self.labels_map = {row['id']: row['cancer'] for _, row in labels_df.iterrows()}
        self.data_path = data_path
        self.target_size = target_size
        self.target_spacing = target_spacing
        self.transform = transform # For potential augmentations later

    def __len__(self):
        return len(self.patient_ids)

    def __getitem__(self, idx):
        patient_id = self.patient_ids[idx]
        label = self.labels_map[patient_id]
        scan_path = os.path.join(self.data_path, patient_id)

        try:
            itk_image = load_itk_image(scan_path)
            image_hu = get_pixels_hu(itk_image)
            spacing = np.array(itk_image.GetSpacing())[::-1] # z, y, x

            # Resample to target spacing
            image_resampled = resample_image(image_hu, spacing, new_spacing=self.target_spacing)

            # Normalize
            image_normalized = normalize(image_resampled)

            # Resize to fixed input size for LCNNET
            image_resized = resize_scan(image_normalized, self.target_size)

            # Add channel dimension and convert to tensor
            scan_tensor = torch.from_numpy(image_resized).float().unsqueeze(0) # (1, D, H, W)
            label_tensor = torch.tensor(label).float() # Float for BCEWithLogitsLoss

            # Apply transforms if any (e.g., augmentation)
            if self.transform:
                 scan_tensor = self.transform(scan_tensor)

            return scan_tensor, label_tensor

        except Exception as e:
            print(f"Error processing patient {patient_id}: {e}. Returning None.")
            # Need robust handling in DataLoader or training loop if this happens
            # For simplicity, we might just skip, but better logging/handling is ideal.
            # Return dummy data or raise error to be caught by dataloader collate_fn
            # Returning None is generally problematic for default collate_fn
            # Let's return zero tensors as placeholders - they should be ignored later if possible
            # A better approach is filtering problematic IDs beforehand
            print(f"Warning: Returning zero tensors for patient {patient_id} due to error.")
            dummy_scan = torch.zeros((1,) + tuple(self.target_size), dtype=torch.float)
            dummy_label = torch.tensor(-1.0).float() # Invalid label
            return dummy_scan, dummy_label

```

```python
# --- PyTorch Model Definitions (DenseNet Style) ---

class _ConvBlock(nn.Sequential):
    def __init__(self, in_channels, out_channels, kernel_size=3, stride=1, dropout_rate=0.2, activation=True):
        super().__init__()
        padding = kernel_size // 2
        self.add_module('conv', nn.Conv3d(in_channels, out_channels, kernel_size=kernel_size, stride=stride, padding=padding, bias=False))
        self.add_module('norm', nn.BatchNorm3d(out_channels))
        if activation:
            self.add_module('relu', nn.ReLU(inplace=True))
        if dropout_rate > 0:
            self.add_module('drop', nn.Dropout3d(dropout_rate))

class _DenseLayer(nn.Module):
    def __init__(self, in_channels, growth_rate, bn_size=4, dropout_rate=0.2):
        super().__init__()
        # Bottleneck layer (optional but standard in DenseNet)
        # self.add_module('norm1', nn.BatchNorm3d(in_channels))
        # self.add_module('relu1', nn.ReLU(inplace=True))
        # self.add_module('conv1', nn.Conv3d(in_channels, bn_size * growth_rate, kernel_size=1, stride=1, bias=False))

        # Main Convolution: Simpler version without bottleneck
        self.norm1 = nn.BatchNorm3d(in_channels)
        self.relu1 = nn.ReLU(inplace=True)
        self.conv1 = nn.Conv3d(in_channels, growth_rate, kernel_size=3, stride=1, padding=1, bias=False)

        self.dropout_rate = dropout_rate
        if self.dropout_rate > 0:
            self.drop = nn.Dropout3d(self.dropout_rate)

    def forward(self, x):
        # Bottleneck version:
        # new_features = super().forward(x)
        # if self.dropout_rate > 0:
        #      new_features = F.dropout3d(new_features, p=self.dropout_rate, training=self.training)
        # return torch.cat([x, new_features], 1)

        # Simpler version:
        out = self.norm1(x)
        out = self.relu1(out)
        new_features = self.conv1(out)
        if self.dropout_rate > 0:
            new_features = self.drop(new_features)
        return torch.cat([x, new_features], 1)


class _DenseBlock(nn.Sequential):
    def __init__(self, num_layers, in_channels, growth_rate, bn_size=4, dropout_rate=0.2):
        super().__init__()
        for i in range(num_layers):
            layer = _DenseLayer(in_channels + i * growth_rate, growth_rate, bn_size, dropout_rate)
            self.add_module(f'denselayer{i+1}', layer)

class _Transition(nn.Sequential):
    def __init__(self, in_channels, out_channels, dropout_rate=0.2):
        super().__init__()
        self.add_module('norm', nn.BatchNorm3d(in_channels))
        self.add_module('relu', nn.ReLU(inplace=True))
        self.add_module('conv', nn.Conv3d(in_channels, out_channels, kernel_size=1, stride=1, bias=False))
        if dropout_rate > 0:
             self.add_module('drop', nn.Dropout3d(dropout_rate)) # Dropout before pooling? Or after? Paper dependent. Let's try after.
        self.add_module('pool', nn.AvgPool3d(kernel_size=2, stride=2))
        # if dropout_rate > 0: # Dropout after pooling
        #      self.add_module('drop_after', nn.Dropout3d(dropout_rate))


class BaseDenseNet3D(nn.Module):
    def __init__(self, growth_rate=16, block_config=(4, 4, 4), num_init_features=32,
                 bn_size=4, compression=0.5, dropout_rate=0.3, num_classes=1):
        super().__init__()

        # Initial convolution
        self.features = nn.Sequential(
            _ConvBlock(1, num_init_features, kernel_size=3, stride=1, dropout_rate=0, activation=True)
            # Optional initial pooling if inputs are larger
            # nn.MaxPool3d(kernel_size=3, stride=2, padding=1),
        )

        # Each denseblock
        num_features = num_init_features
        for i, num_layers in enumerate(block_config):
            block = _DenseBlock(num_layers=num_layers, in_channels=num_features,
                                growth_rate=growth_rate, bn_size=bn_size, dropout_rate=dropout_rate)
            self.features.add_module(f'denseblock{i+1}', block)
            num_features = num_features + num_layers * growth_rate
            if i != len(block_config) - 1: # Add transition except for last block
                num_out_features = int(num_features * compression)
                trans = _Transition(in_channels=num_features, out_channels=num_out_features, dropout_rate=dropout_rate)
                self.features.add_module(f'transition{i+1}', trans)
                num_features = num_out_features

        # Final batch norm
        self.features.add_module('norm_final', nn.BatchNorm3d(num_features))
        self.features.add_module('relu_final', nn.ReLU(inplace=True))

        # Linear layer
        self.avgpool = nn.AdaptiveAvgPool3d((1, 1, 1))
        self.classifier = nn.Linear(num_features, num_classes)

        # Official init from torch repo.
        for m in self.modules():
            if isinstance(m, nn.Conv3d):
                nn.init.kaiming_normal_(m.weight)
            elif isinstance(m, nn.BatchNorm3d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.Linear):
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        features = self.features(x)
        out = self.avgpool(features)
        out = torch.flatten(out, 1)
        out = self.classifier(out)
        return out
```

```python
# --- Training and Evaluation Functions ---

def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, device, epochs, model_name):
    best_val_loss = float('inf')
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    model_save_path = os.path.join(OUTPUT_DIR, f'{model_name}_best.pth')

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        correct_train = 0
        total_train = 0

        train_pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} [Train]")
        for inputs, labels in train_pbar:
            # Skip potentially problematic batches from dataset __getitem__ error
            if torch.any(labels == -1.0):
                print("Skipping a batch due to data loading error.")
                continue

            inputs, labels = inputs.to(device), labels.to(device).unsqueeze(1) # Ensure labels are [B, 1]

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)

            # Accuracy calculation
            preds = torch.sigmoid(outputs) > 0.5
            correct_train += (preds == labels.byte()).sum().item()
            total_train += labels.size(0)

            train_pbar.set_postfix({'loss': loss.item(), 'acc': correct_train / total_train if total_train > 0 else 0})


        epoch_loss = running_loss / len(train_loader.dataset)
        epoch_acc = correct_train / total_train if total_train > 0 else 0
        history['train_loss'].append(epoch_loss)
        history['train_acc'].append(epoch_acc)


        # Validation
        model.eval()
        val_loss = 0.0
        correct_val = 0
        total_val = 0
        all_labels_val = []
        all_preds_val = []

        val_pbar = tqdm(val_loader, desc=f"Epoch {epoch+1}/{epochs} [Val]")
        with torch.no_grad():
            for inputs, labels in val_pbar:
                 if torch.any(labels == -1.0): continue # Skip bad batches

                 inputs, labels = inputs.to(device), labels.to(device).unsqueeze(1)
                 outputs = model(inputs)
                 loss = criterion(outputs, labels)
                 val_loss += loss.item() * inputs.size(0)

                 preds_prob = torch.sigmoid(outputs)
                 preds = preds_prob > 0.5
                 correct_val += (preds == labels.byte()).sum().item()
                 total_val += labels.size(0)
                 
                 all_labels_val.extend(labels.cpu().numpy().flatten())
                 all_preds_val.extend(preds.cpu().numpy().flatten())

                 val_pbar.set_postfix({'loss': loss.item(), 'acc': correct_val / total_val if total_val > 0 else 0})


        epoch_val_loss = val_loss / len(val_loader.dataset)
        epoch_val_acc = correct_val / total_val if total_val > 0 else 0
        history['val_loss'].append(epoch_val_loss)
        history['val_acc'].append(epoch_val_acc)


        print(f"Epoch {epoch+1}: Train Loss: {epoch_loss:.4f}, Train Acc: {epoch_acc:.4f} | Val Loss: {epoch_val_loss:.4f}, Val Acc: {epoch_val_acc:.4f}")

        # Save best model
        if epoch_val_loss < best_val_loss:
            print(f"Validation loss decreased ({best_val_loss:.4f} --> {epoch_val_loss:.4f}). Saving model...")
            best_val_loss = epoch_val_loss
            torch.save(model.state_dict(), model_save_path)

        # LR Scheduler Step
        if scheduler:
            scheduler.step(epoch_val_loss) # ReduceLROnPlateau needs the metric

    print(f"Training finished. Best model saved to {model_save_path}")
    return history
```

```python
def evaluate_model(model, data_loader, criterion, device, model_name):
    model.eval()
    running_loss = 0.0
    all_labels = []
    all_preds = []
    all_probs = []

    with torch.no_grad():
        for inputs, labels in tqdm(data_loader, desc=f"Evaluating {model_name}"):
            if torch.any(labels == -1.0): continue
            inputs, labels = inputs.to(device), labels.to(device).unsqueeze(1)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            running_loss += loss.item() * inputs.size(0)

            probs = torch.sigmoid(outputs)
            preds = probs > 0.5

            all_labels.extend(labels.cpu().numpy().flatten())
            all_preds.extend(preds.cpu().numpy().flatten())
            all_probs.extend(probs.cpu().numpy().flatten())

    final_loss = running_loss / len(data_loader.dataset)
    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, zero_division=0)
    recall = recall_score(all_labels, all_preds, zero_division=0)
    f1 = f1_score(all_labels, all_preds, zero_division=0)
    try:
        auc = roc_auc_score(all_labels, all_probs)
    except ValueError: # Handle cases with only one class in labels
        auc = -1.0
        print("Warning: AUC calculation failed (likely only one class present in evaluation set).")


    print(f"\n--- Evaluation Results ({model_name}) ---")
    print(f"Loss: {final_loss:.4f}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"AUC: {auc:.4f}")

    print("\nClassification Report:")
    print(classification_report(all_labels, all_preds, target_names=['Class 0', 'Class 1'], zero_division=0))

    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(7, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Pred 0', 'Pred 1'], yticklabels=['True 0', 'True 1'])
    plt.title(f'Confusion Matrix ({model_name})')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.savefig(os.path.join(OUTPUT_DIR, f'confusion_matrix_{model_name}.png'))
    plt.show()

def plot_history(history, model_name):
    epochs_range = range(len(history['train_loss']))
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, history['train_acc'], label='Training Accuracy')
    plt.plot(epochs_range, history['val_acc'], label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title(f'Accuracy ({model_name})')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, history['train_loss'], label='Training Loss')
    plt.plot(epochs_range, history['val_loss'], label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title(f'Loss ({model_name})')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')

    plt.savefig(os.path.join(OUTPUT_DIR, f'training_history_{model_name}.png'))
    plt.show()
```

```python
import matplotlib.pyplot as plt

def show_scan_slices(volume, title="Scan", slice_indices=None, cmap='gray'):
    """Displays axial, coronal, and sagittal slices of a 3D volume."""
    if slice_indices is None:
        slice_z = volume.shape[0] // 2
        slice_y = volume.shape[1] // 2
        slice_x = volume.shape[2] // 2
    else:
        slice_z, slice_y, slice_x = slice_indices

    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    fig.suptitle(title, fontsize=16)

    axes[0].imshow(volume[slice_z, :, :], cmap=cmap)
    axes[0].set_title('Axial (Z)')

    axes[1].imshow(volume[:, slice_y, :], cmap=cmap)
    axes[1].set_title('Coronal (Y)')

    axes[2].imshow(volume[:, :, slice_x], cmap=cmap)
    axes[2].set_title('Sagittal (X)')

    for ax in axes:
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# --- Clean up after training on one subset ---
del X_ldnnet, y_ldnnet
del train_dataset_ld, val_dataset_ld
del train_loader_ld, val_loader_ld

import gc
gc.collect()

import torch
torch.cuda.empty_cache()

# --- 1. LDNNET: Load and Preprocess LUNA16 Patches ---
LUNA16_DATA_PATH = '/kaggle/input/luna-16/Dataset/subset9'
print("--- 1. Loading and Preprocessing LUNA16 for LDNNET ---")
luna_annotations = pd.read_csv(LUNA16_ANNOTATIONS_PATH)
luna_files = []
for i in range(10):
    subset_path = os.path.join(LUNA16_DATA_PATH, f'subset{i}')
    if os.path.isdir(subset_path):
        luna_files.extend(glob(os.path.join(subset_path, '*.mhd')))
    else:
        print(f"Warning: Subset path not found: {subset_path}")

print(f"Found {len(luna_files)} LUNA16 scan files.")
luna_file_map = {os.path.basename(f).replace('.mhd', ''): f for f in luna_files}
luna_annotations = luna_annotations[luna_annotations['seriesuid'].isin(luna_file_map.keys())]
print(f"Processing {len(luna_annotations)} candidates from available scans.")

X_ldnnet = []
y_ldnnet = []
processed_scans_cache = {} # Cache processed scans
processed_scan_metadata_cache = {} # Cache origin/spacing

# Limiting for faster dev run - REMOVE FOR FULL TRAINING
# luna_annotations = luna_annotations.sample(n=500, random_state=42)
# --- Subsampling for Class Balance (2:1 Ratio) ---

class_1 = luna_annotations[luna_annotations['class'] == 1]
class_0 = luna_annotations[luna_annotations['class'] == 0]

# Number of class 1 candidates
n_class_1 = len(class_1)

# Sample 2x that many from class 0
class_0_sampled = class_0.sample(n_class_1, random_state=42)

# Combine and shuffle
luna_annotations = pd.concat([class_1, class_0_sampled]).sample(frac=1.0, random_state=42).reset_index(drop=True)

print(f"Subsampled annotations: Class 0 = {len(class_0_sampled)}, Class 1 = {n_class_1}, Total = {len(luna_annotations)}")

# print("--- WARNING: Running on a limited subset of LUNA16 candidates ---")


start_time = time.time()
for i, row in tqdm(luna_annotations.iterrows(), total=len(luna_annotations), desc="Extracting LUNA16 Patches"):
    seriesuid = row['seriesuid']
    world_coord = np.array([row['coordZ'], row['coordY'], row['coordX']])
    label = int(row['class'])

    if seriesuid in processed_scans_cache:
        image_normalized = processed_scans_cache[seriesuid]
        origin, spacing = processed_scan_metadata_cache[seriesuid]
    else:
        if seriesuid not in luna_file_map: continue
        file_path = luna_file_map[seriesuid]
        try:
            itk_image = load_itk_image(file_path)
            origin = np.array(itk_image.GetOrigin())[::-1] # zyx
            spacing = np.array(itk_image.GetSpacing())[::-1] # zyx
            image_hu = get_pixels_hu(itk_image)
            image_resampled = resample_image(image_hu, spacing, new_spacing=TARGET_SPACING)
            image_normalized = normalize(image_resampled)
            processed_scans_cache[seriesuid] = image_normalized
            processed_scan_metadata_cache[seriesuid] = (origin, spacing) # Store metadata
        except Exception as e:
            print(f"Error processing LUNA16 scan {seriesuid}: {e}")
            processed_scans_cache[seriesuid] = None # Mark as failed
            processed_scan_metadata_cache[seriesuid] = (None, None)
            continue

    if image_normalized is None or origin is None: # Skip if scan processing failed
        continue

    # Convert world to voxel on original, then scale to resampled space
    voxel_coord_orig = world_to_voxel(world_coord, origin, spacing)
    resize_factor_coord = spacing / TARGET_SPACING # How original spacing relates to target
    voxel_coord_resampled = np.round(np.array(voxel_coord_orig) * resize_factor_coord).astype(int) # Use np.array()
    voxel_coord_resampled = np.minimum(voxel_coord_resampled, np.array(image_normalized.shape) - 1)
    voxel_coord_resampled = np.maximum(voxel_coord_resampled, np.array([0,0,0]))

    patch = extract_patch(image_normalized, voxel_coord_resampled, PATCH_SIZE)

    if patch is not None and patch.shape == PATCH_SIZE:
        X_ldnnet.append(patch)
        y_ldnnet.append(label)

print(f"LUNA16 patch extraction took {time.time() - start_time:.2f}s")
del processed_scans_cache # Free up memory
del processed_scan_metadata_cache

if not X_ldnnet:
    raise ValueError("No LDNNET patches extracted. Check paths and preprocessing.")

X_ldnnet = np.array(X_ldnnet)
y_ldnnet = np.array(y_ldnnet)

print(f"LDNNET Data: Patches={X_ldnnet.shape}, Labels={y_ldnnet.shape}")
print(f"LDNNET Class distribution: {np.bincount(y_ldnnet)}")

# Split LDNNET data
X_train_ld, X_val_ld, y_train_ld, y_val_ld = train_test_split(
    X_ldnnet, y_ldnnet, test_size=VAL_SPLIT_RATIO, random_state=42, stratify=y_ldnnet
)

# Create Datasets and DataLoaders for LDNNET
train_dataset_ld = LUNAPatchDataset(X_train_ld, y_train_ld)
val_dataset_ld = LUNAPatchDataset(X_val_ld, y_val_ld)

train_loader_ld = DataLoader(train_dataset_ld, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader_ld = DataLoader(val_dataset_ld, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

current_subset_index = 1
print(f"\n--- Training LDNNET on subset {current_subset_index} ---")

# Load model (with previous weights if continuing training)
previous_model_path = os.path.join(OUTPUT_DIR, 'LDNNET_best.pth')
ldnnet_model = setup_ldnnet_model(load_path=previous_model_path if current_subset_index > 0 else None)

optimizer_ld = optim.Adam(ldnnet_model.parameters(), lr=LEARNING_RATE)
criterion_ld = nn.BCEWithLogitsLoss()
scheduler_ld = optim.lr_scheduler.ReduceLROnPlateau(optimizer_ld, 'min', patience=5, factor=0.2)

# Train
history_ld = train_model(
    ldnnet_model, 
    train_loader_ld, 
    val_loader_ld, 
    criterion_ld, 
    optimizer_ld, 
    scheduler_ld, 
    DEVICE, 
    EPOCHS_LDNNET, 
    "LDNNET"
)

plot_history(history_ld, f"LDNNET_subset{current_subset_index}")
# Save latest weights after training this subset
torch.save(ldnnet_model.state_dict(), os.path.join(OUTPUT_DIR, f'LDNNET_subset{current_subset_index}.pth'))

evaluate_model(ldnnet_model, val_loader_ld, criterion_ld, DEVICE, "LDNNET_Validation")
```
